```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return x**2 - 4*x + 4

# Generate x values between -10 and 10
x = np.linspace(-10, 10, 500)

# Calculate corresponding y values
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = x^2 - 4x + 4$", color="blue")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # x-axis
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  # y-axis

# Add labels, title, and legend
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.title("Plot of $f(x) = x^2 - 4x + 4$", fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 500)

# Calculate y values for sine and cosine
sin_y = np.sin(x)
cos_y = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, sin_y, label=r"$\sin(x)$", color="red", linestyle="-", marker=None)
plt.plot(x, cos_y, label=r"$\cos(x)$", color="blue", linestyle="--", marker=None)

# Add labels, title, and legend
plt.xlabel("x (radians)", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("Sine and Cosine Functions", fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
# Define the functions
def cubic(x):
    return x**3

def sine(x):
    return np.sin(x)

def exponential(x):
    return np.exp(x)

def logarithm(x):
    return np.log(x + 1)

# Define x ranges for each function
x_cubic = np.linspace(-10, 10, 500)
x_sine = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x_exponential = np.linspace(-2, 2, 500)
x_logarithm = np.linspace(0, 10, 500)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Top-left: Cubic function
axes[0, 0].plot(x_cubic, cubic(x_cubic), color="purple")
axes[0, 0].set_title("$f(x) = x^3$")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("f(x)")
axes[0, 0].grid(alpha=0.3)

# Top-right: Sine function
axes[0, 1].plot(x_sine, sine(x_sine), color="green")
axes[0, 1].set_title("$f(x) = \sin(x)$")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("f(x)")
axes[0, 1].grid(alpha=0.3)

# Bottom-left: Exponential function
axes[1, 0].plot(x_exponential, exponential(x_exponential), color="blue")
axes[1, 0].set_title("$f(x) = e^x$")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("f(x)")
axes[1, 0].grid(alpha=0.3)

# Bottom-right: Logarithmic function
axes[1, 1].plot(x_logarithm, logarithm(x_logarithm), color="red")
axes[1, 1].set_title("$f(x) = \log(x+1)$")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("f(x)")
axes[1, 1].grid(alpha=0.3)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Generate random x and y values
np.random.seed(42)  # For reproducibility
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=np.random.rand(100), cmap="viridis", marker="o", edgecolor="black")

# Add labels, title, and grid
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("Scatter Plot of Random Points", fontsize=14)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
# Generate a random dataset from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color="blue", alpha=0.7, edgecolor="black")

# Add labels, title, and grid
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Histogram of Normally Distributed Data", fontsize=14)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Define the function f(x,y)
def f(x,y):
    return np.cos(x**2 + y**2)

# Create a gird of x and y
x = np.linspace(-5,5, 1000)
y = np.linspace(-5,5, 1000)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

#Create the 3D plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

#Plot the surface
surface = ax.plot_surface(X,Y,Z , cmap ='viridis', edgecolor = 'none')

#Add colorbar
colorbar = fig.colorbar(surface, ax=ax, shrink=0.5, aspect = 10)

# Set labels and title
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.set_zlabel("Z-axis", fontsize=12)
ax.set_title("3D Surface Plot of $f(x, y) = \cos(x^2 + y^2)$", fontsize=14)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'green', 'purple', 'orange']

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors, edgecolor='black')

# Add title and axis labels
plt.title("Sales Data for Different Products", fontsize=14)
plt.xlabel("Products", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Sample data for the stacked bar chart
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [30, 40, 35, 50]
category_b = [20, 25, 30, 35]
category_c = [15, 20, 25, 30]

# Create the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(time_periods, category_a, label='Category A', color='skyblue')
plt.bar(time_periods, category_b, bottom=category_a, label='Category B', color='salmon')
plt.bar(time_periods, category_c, bottom=np.array(category_a) + np.array(category_b), label='Category C', color='lightgreen')

# Add title, axis labels, and legend
plt.title("Stacked Bar Chart of Category Contributions", fontsize=14)
plt.xlabel("Time Periods", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend(title="Categories", fontsize=10)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
