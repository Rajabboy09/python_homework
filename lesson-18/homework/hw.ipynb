```python
#1
numeric_columns = df_iris.select_dtypes(include=["number"])
mean_values = numeric_columns[["sepalLength", "sepalWidth","petalLength","petalWidth"]].mean()
median_values = numeric_columns[["sepalLength", "sepalWidth","petalLength","petalWidth"]].median()
std_values = numeric_columns[["sepalLength", "sepalWidth","petalLength","petalWidth"]].std()
print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nStandard Deviation values:\n", std_values)
```

```python
print("Sum ages of passangers: ", df_titanic["Age"].sum())
print("Minimum age of passangers: " , df_titanic["Age"].min())
print("Maximum age pf passangers: ", df_titanic["Age"].max())
```

```python
director_max_likes = df_movies.groupby('director_name')['director_facebook_likes'].sum().idxmax()

# Display the result
max_likes_value = df_movies.groupby('director_name')['director_facebook_likes'].sum().max()
print(f"Director with the highest total 'director_facebook_likes': {director_max_likes} with {max_likes_value} likes.")

df_longest_movies = df_movies[['director_name', 'duration']].sort_values(by='duration', ascending=False).head(5)
print("\nThe 5 longest movies and their respective directors:")
print(df_longest_movies)
```
